```python
con_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(con_img)
```

```python
img_g=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(img_g)
```

```python
img_g
```

```python
face_cas=cv2.CascadeClassifier(r'C:\DL Projects\Face Detection\haarcascade_frontalface_default.xml')
```

```python
face_cas
```

```python
face=face_cas.detectMultiScale(img_g,minNeighbors=6,scaleFactor=1.1)
```

```python
face
```

```python
len(face)
```

```python
for (x,y,w,h) in face:
    cv2.rectangle(con_img,pt1=(x,y),pt2=(x+w,y+h),color=(255,255,0),thickness=4)
```

HOW TO MAKE video FACE DETECTION(11-07-2023)

```python
video=cv2.VideoCapture(r'C:\DL Projects\Face Detection\video.mp4')
while True:
    suc,frame=video.read()  #reads frames at every execution
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    # print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,0),3)
    cv2.imshow('video',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```

WEBCAM FACE DETECTION

```python
video=cv2.VideoCapture(0)
while True:
    suc,frame=video.read()
    gray=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    faces=face_cas.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=10)
    # print(len(faces))
    for (x,y,w,h) in faces:
        cv2.rectangle(frame,(x,y),(x+w,y+h),(255,255,0),3)

    cv2.imshow('Webcam',frame)
    if cv2.waitKey(1) & 0XFF==ord('q'):
        break
video.release()
cv2.destroyAllWindows()
```
